```python
### 1) python function to find factorial

def factorial(num):
  fact = 1
  for i in range(1,num+1):
    fact*=i
  return fact
```

```python
### 3) calculate number of digits
def num_digits(num):
  _num=num
  digits_count=0
  while _num>0:
    digits_count+=1
    _num//=10 
  return digits_count
```

```python
num_digits(int(input('Enter number - ')))
```

```python
### 4) product of factorial of all natural numbers till a number

def product_of_factorials(num):
  prod=1
  for i in range(1,num+1):
    prod*=factorial(i) # using factorial function in part 1
  return prod
```

```python
product_of_factorials(int(input("Enter number - ")))
```

```python
### 5) sum of squares of all supplied numbers

def sums_of_squares(*numbers):
  _sum=0
  for i in numbers:
    _sum+=i**2
  return _sum
```

```python
sums_of_squares(1,2,3)
```

```python
### 6) creating a basic calculator

def calculator(a,b,c):
  if c=="+":
    return a+b
  elif c=="-":
    return a-b
  elif c=="*":
    return a*b
  elif c=="/":
    return a/b
```

```python
calculator(float(input('Enter first number - ')), float(input('Enter second number - ')), input("Enter function +,-,*, or / - "))
```

```python
### 7) "I CAN" & "I WILL"

def can_and_will(num):
  for i in range(num):
    if i%2==0:
      print('I CAN')
    else:
      print('I WILL')
```

```python
can_and_will(int(input('Enter number - ')))
```

```python
### 8) pairing up socks
def matching_socks(sock_list):
  sock_set=set(sock_list)
  pairs=0
  for i in sock_set:
    pairs+=sock_list.count(i)//2
  return pairs
```

```python
matching_socks([1,2,3,1,2,2])
```

```python
### 9) check for consecutive sequence
def isConsecutive(list1,list2):
  list_combo=list1 + list2
  list_combo.sort()
  return list_combo==[i for i in range(min(list_combo),max(list_combo)+1,1)]
```

```python
isConsecutive([0,1,2],[5,6,4])
```

```python
### 10) Calculate Total profits

def profit(prices):
  return round((prices['sell_price'] - prices['cost_price'])*prices['inventory'])
```

```python
profit({'cost_price':float(input('Enter cost price - ')), 'sell_price':float(input('Enter selling price - ')), 'inventory':int(input('Enter number of units available in inventory - '))})
```
